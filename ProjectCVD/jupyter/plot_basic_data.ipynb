```python
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

save_path = "graphs"
os.makedirs(save_path, exist_ok=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
df = pd.read_csv('data/HeartFailureDataset.csv')
df["age_years"] = (df["age"] / 365).astype(int)
```

```python
def save_and_show(name): # Сохрняем графики в папку в формате PNG
    plt.tight_layout()
    plt.savefig(f"graphs/{name}.png", dpi=300)
    plt.show()
    plt.close()
```

<h1>Базовые графики</h1>

```python
plt.figure(figsize=(8,5))

sns.histplot(df["age_years"], kde=True, bins=30, color='green')

plt.title("Распределение возраста")
plt.xlabel("Возраст")
plt.ylabel("Количество")

save_and_show("age_distribution")
```

```python
df["gender_label"] = df["gender"].map({1: "Женщина", 2: "Мужчина"})

plt.figure(figsize=(8,5))

sns.countplot(x="gender_label", data=df, palette=["#5BCFFB", "#F5ABB9"])

plt.title("Распределение по полу")
plt.xlabel("Пол")
plt.ylabel("Количество людей")

save_and_show("gender_distribution")
```

```python
plt.figure(figsize=(6,6))

gender_counts = df["gender_label"].value_counts()

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["#F5ABB9", "#5BCFFB"]
)

plt.title("Процентное распределение пола")

plt.axis("equal")

save_and_show("gender_pie_chart")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["height"], kde=True, bins=30)

plt.title("Распределение роста")
plt.xlabel("Рост")
plt.ylabel("Количество")

save_and_show("height_distribution")
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["weight"], kde=True, bins=30, color='red')

plt.title("Распределение веса")
plt.xlabel("Вес")
plt.ylabel("Количество")

save_and_show("weight_distribution")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="cholesterol", data=df, palette=["#4CAF50", "#FFC107", "#F44336"])

plt.title("Распределение уровня холестерина")
plt.xlabel("Уровень холестерина\n(1 = нормальный, 2 = выше нормы, 3 = значительно выше нормы)")
plt.ylabel("Количество")

save_and_show("cholesterol_distribution")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="gluc", data=df, palette=["#4CAF50", "#FFC107", "#F44336"])

plt.title("Распределение уровня глюкозы")
plt.xlabel("Уровень глюкозы\n(1 = нормальный, 2 = выше нормы, 3 = значительно выше нормы)")
plt.ylabel("Количество")

save_and_show("glucose_distribution")
```

```python
def add_labels(ax):
    for p in ax.patches:
        ax.annotate(
            f'{int(p.get_height())}',
            (p.get_x() + p.get_width() / 2, p.get_height()),
            ha='center',
            va='bottom'
        )
```

```python
plt.figure(figsize=(7,5))

ax = sns.countplot(
    x="smoke",
    data=df,
    palette=["#4CAF50", "#F44336"]
)

plt.title("Курение")
plt.xlabel("Курит")
plt.ylabel("Количество людей")

ax.set_xticklabels(["Нет", "Да"])

add_labels(ax)

save_and_show("smoking_distribution")
```
